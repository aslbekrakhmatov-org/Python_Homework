##### task1

```python
import pandas as pd
df_titanic = pd.read_excel("../data/titanic.xlsx")
def check_survived(x):
    return x[x["Survived"]==1].copy() 
df_titanic.pipe(check_survived)
def fill_AgeNaNs(x):
    x.loc[:, "Age"] = x["Age"].fillna(x["Age"].mean())
    return x
df_titanic.pipe(fill_AgeNaNs)
def Fare_per_age(x):
    x.loc[:, "Fare_per_Age"]=x['Fare']/x['Age']
    return x
df_titanic = (
    df_titanic.pipe(check_survived)
    .pipe(fill_AgeNaNs)
    .pipe(Fare_per_age)
)
df_titanic
```

##### task2

```python
df_flights = pd.read_parquet("../data/flights")
def filter_by_delay(df):
    return df[pd.to_numeric(df["DepDelay"], errors="coerce")>30].copy()

def Delay_per_Hour(df):
    df.loc[:, "Delay_Per_Hour"]= pd.to_numeric(df["DepDelay"], errors="coerce")/pd.to_numeric(df["Duration"], errors="coerce")

df_flights=(
    df_flights.pipe(filter_by_delay)
    .pipe(Delay_per_Hour)
)
df_flights
```
